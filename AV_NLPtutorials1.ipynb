**Text Preprocessing**

1.Noise Removal:
removing stop words like is,am,the,of or urls or links or social media hastags, punctuations, industry specific words is called noise removal

```python
noise_list = ["is", "a", "this", "..."]
def _remove_noise(input_text):
    words = input_text.split()
    noise_free_words = [word for word in words if word not in noise_list]
    noise_free_text = " ".join(noise_free_words)
    return noise_free_text

print(_remove_noise("this is a sample text"))
```

2.Lexicon Normalization: type in which we reduce different representation of a word to its base/stem

technqiues mostly used for this are:
1.stemming
2.lemmatization

**Stemming:** here we remove the suffixes like "ing,"ly","es","s" etc from a word

**Lemmatization:** is step by step procedure of obtaining the root form of the word

Text Data Cleaning

```python
import string
string.punctuation

#defining the function to remove punctuation

def remove_punctuation(text):
  pfree = "".join([i for i in text if i not in string.punctuation])
  return pfree
```

```python
#storing punctuation free text
df['clean_msg'] = df['v2'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
df.drop(columns = 'v2',axis=1)
```

**Text Lowering:**

```python
df['msg_lower'] = df['clean_msg'].apply(lambda x: x.lower())
```

**Tokenisation**: splitting of text into smaller units. we either use sentence tokenisation or word tokenisation based on problem statement

```python
import re
def tokenisation(text):
  tokens = re.split('W+',text)
  return tokens

#applying functions to the column
df['msg_tokenised'] = df['msg_lower'].apply(lambda x: tokenisation(x))
```

```python
df.head()
```

**Stop Words:** removing of commonly used words that do not affect the sentence even after their removal

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords[:10]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Define the remove_sw function
def remove_sw(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]
    return ' '.join(filtered_words)

# Apply the remove_sw function to create the 'no_stopwords' column
df['no_stopwords'] = df['msg_lower'].apply(remove_sw)
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
#defining a function for stemming
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text

df['msg_stemmed']=df['no_stopwords'].apply(lambda x: stemming(x))
```

```python
df.head()
```

```python
df.drop(columns=['msg_stemmed'],axis=1)
```

```python
nltk.download('punkt')
porter_stemmer = PorterStemmer()

# Define the stemming function
def stemming(text):
    words = nltk.word_tokenize(text)
    stemmed_words = [porter_stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Apply the stemming function to create the 'msg_stemmed' column
df['msg_stemmed'] = df['no_stopwords'].apply(stemming)
```

```python
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
#defining the function for lemmatization
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
df['msg_lemmatized']=df['no_stopwords'].apply(lambda x:lemmatizer(x))

df.drop(columns=['msg_lemmatized'],axis=1)
```
